```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# 2. Plot stacked bar chart
ax = ct.plot(kind='bar', stacked=True, colormap='Set2')

# 3. Add percentage labels
for i, row in ct.iterrows():
    cum_sum = 0
    for j, val in enumerate(row):
        plt.text(i, cum_sum + val/2, f'{val:.1f}%', ha='center', va='center', fontsize=10)
        cum_sum += val

# 4. Format the plot
plt.title("Churn Percentage by SeniorCitizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.xticks(ticks=[0, 1], labels=["No", "Yes"], rotation=0)
plt.legend(title="Churn", loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Customer Churn.csv')
df
```

```python
df.columns.values
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplot grid
n_cols = 3  # Number of plots per row
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate required rows

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 12))
axes = axes.flatten()  # Flatten 2D array of axes for easy indexing

# Loop through columns and plot each
for i, col in enumerate(columns):
    ax = axes[i]
    sns.countplot(x=col, data=df, ax=ax, hue=df["Churn"])
    ax.set_title(f'{col} Count')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.bar_label(ax.containers[0])
    ax.tick_params(axis='x', rotation=30)

# Remove any unused subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#Here’s a brief summary of the visualized data based on your subplots:

🔹 Customers without internet service consistently show low engagement across features like streaming, security, and tech support.

🔹 Churn is higher among those without optional services or with Fiber optic internet, suggesting these may correlate with dissatisfaction or higher risk.

🔹 Binary services like PhoneService and MultipleLines show similar churn patterns, though most customers seem to opt for having phone service.
